```python
import pandas
```

```python
max_level = 4
cantrips = ['Druidcraft', 'Mold Earth', 'Shillelagh']
infilename = 'druid_spells.csv'
outfilename = 'druid_spells.html'
```

```python
df = pandas.read_csv(infilename).set_index("title")
df.head()
```

```python
df = df.loc[(df['level'] > 0) | (df.index.isin(cantrips))]
df = df.loc[df['level'] <= max_level]
df.tail()
```

```python
df = df.loc[~df['spell lists'].str.contains("Druid \(Optional\)")]
df.head()
```

```python
def to_html(spell):
    result = ""
    result += '<div class="spell">\n'
    result += f'<h1>{spell.name}</h1>\n'
    result += f'Level: <b>{spell["level"]}</b>; Casting Time: <b>{spell["casting time"]}</b>; Range: <b>{spell["range"]}</b>; Components: <b>{spell["components"]}</b>; Duration <b>{spell["duration"]}</b><br>\n'
    result += spell["description"]+"\n"
    result += f'Source: <em>{spell["source"]}</em>\n'
    result += f'</div>'
    return result
```

```python
header = f'''
<DOCTYPE html>
<html><head>
  <title>Spell list from {infilename}</title>
  <style>
    .spell {
      break-inside: avoid;
    }
  </style>
</head>
<body>
'''
footer = f'</body></html>'
with open(outfilename, "w") as outfile:
    outfile.write(header)
    for index, spell in df.iterrows():
        outfile.write(to_html(spell))
    outfile.write(footer)
```
